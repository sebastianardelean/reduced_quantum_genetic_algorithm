```python
import math
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
NO_OF_QUBITS_FITNESS = 8
NO_OF_QUBITS_INDIVIDUAL = 4
POPULATION_SIZE = 2**NO_OF_QUBITS_INDIVIDUAL
NO_OF_MAX_GROVER_ITERATIONS = int(math.sqrt(2**NO_OF_QUBITS_INDIVIDUAL-1))
NO_OF_QUBITS_CARRY = 2 * NO_OF_MAX_GROVER_ITERATIONS + 1
NO_OF_GENERATIONS = 8
```

# Number of qubits

```python
VALIDITY_QUBIT = 1 
ORACLE_WS_QUBIT=1
total_number_of_qubits = NO_OF_QUBITS_INDIVIDUAL + 2*NO_OF_QUBITS_FITNESS+ VALIDITY_QUBIT+ ORACLE_WS_QUBIT+NO_OF_QUBITS_CARRY
print(f"Total number of qubits {total_number_of_qubits}")
```

```python
def total_number_of_qubits(individual_size):
    NO_OF_QUBITS_FITNESS = 8
    NO_OF_QUBITS_INDIVIDUAL = individual_size
    POPULATION_SIZE = 2**NO_OF_QUBITS_INDIVIDUAL
    NO_OF_MAX_GROVER_ITERATIONS = int(math.sqrt(2**NO_OF_QUBITS_INDIVIDUAL-1))
    NO_OF_QUBITS_CARRY = 2 * NO_OF_MAX_GROVER_ITERATIONS + 1
    NO_OF_GENERATIONS = 8
    VALIDITY_QUBIT = 1 
    ORACLE_WS_QUBIT=1
    total_number_of_qubits = NO_OF_QUBITS_INDIVIDUAL + 2*NO_OF_QUBITS_FITNESS+ VALIDITY_QUBIT+ ORACLE_WS_QUBIT+NO_OF_QUBITS_CARRY
    return total_number_of_qubits


individual_size = [4,8,12,16,20,24]
total_qubits = list(map(lambda x:total_number_of_qubits(x),individual_size))

fig, ax = plt.subplots()
ax.plot(individual_size, total_qubits,'o--', label="Total number of qubits used", color = "black",linewidth=3,markersize=10)
ax.legend()
ax.grid()
ax.set_xlabel("Number of qubits used for individual representation",fontsize=16)
ax.set_ylabel("Total number of qubits",fontsize=16)
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = "14"
plt.xticks(individual_size)
plt.show()
```

# Number of gates

```python
H_GATES = NO_OF_QUBITS_INDIVIDUAL+1+NO_OF_MAX_GROVER_ITERATIONS*(2+2*NO_OF_QUBITS_FITNESS+3)
X_GATES = POPULATION_SIZE*2*NO_OF_QUBITS_INDIVIDUAL+NO_OF_MAX_GROVER_ITERATIONS*(2+3*NO_OF_QUBITS_FITNESS)
CX_GATES = 2*NO_OF_MAX_GROVER_ITERATIONS*(2*8+2*8+1)
CCX_GATES = 2*NO_OF_MAX_GROVER_ITERATIONS*(8+8)
MCT_GATES = POPULATION_SIZE*NO_OF_QUBITS_FITNESS+NO_OF_MAX_GROVER_ITERATIONS*2
TOTAL = H_GATES+X_GATES+CX_GATES+CCX_GATES+MCT_GATES

print(f"H:{H_GATES}\tX:{X_GATES}\tCX:{CX_GATES}\tCCX:{CCX_GATES}\tMCT:{MCT_GATES}")
```

```python
def f(no_of_qubits_individual, grover_iterations):
    NO_OF_QUBITS_FITNESS = 8
    NO_OF_QUBITS_INDIVIDUAL = no_of_qubits_individual
    POPULATION_SIZE = 2**NO_OF_QUBITS_INDIVIDUAL
    NO_OF_MAX_GROVER_ITERATIONS = grover_iterations#int(math.sqrt(2**NO_OF_QUBITS_INDIVIDUAL-1))
    NO_OF_QUBITS_CARRY = 2 * NO_OF_MAX_GROVER_ITERATIONS + 1
    NO_OF_GENERATIONS = 8
    
    H_GATES = NO_OF_QUBITS_INDIVIDUAL+1+NO_OF_MAX_GROVER_ITERATIONS*(2+2*NO_OF_QUBITS_FITNESS+3)
    X_GATES = POPULATION_SIZE*2*NO_OF_QUBITS_INDIVIDUAL+NO_OF_MAX_GROVER_ITERATIONS*(2+3*NO_OF_QUBITS_FITNESS)
    CX_GATES = 2*NO_OF_MAX_GROVER_ITERATIONS*(2*8+2*8+1)
    CCX_GATES = 2*NO_OF_MAX_GROVER_ITERATIONS*(8+8)
    MCT_GATES = POPULATION_SIZE*NO_OF_QUBITS_FITNESS+NO_OF_MAX_GROVER_ITERATIONS*2
    TOTAL = H_GATES+X_GATES+CX_GATES+CCX_GATES+MCT_GATES
    TOTAL = H_GATES+X_GATES+CX_GATES+CCX_GATES+MCT_GATES
    return TOTAL


x = [4, 8,12,16,20,24]
y = list(map(lambda x:int(math.sqrt(2**x-1)), x))
xy= list(zip(x,y))
z= list(map(lambda x:f(x[0],x[1]),xy))
print(y)
print(z)
fig,ax1 = plt.subplots()

ax1.set_xlabel("Number of qubits in individual register",fontsize=16)
ax1.set_ylabel("Number of Grover Iterations", fontsize=16)
#ax1.set_yscale('log')
ax1.plot(x,y,'ko--',label="Number of Grover Iterations")
ax1.tick_params(axis='y')
ax1.legend(loc=0)
ax1.grid()
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_yscale('log')

ax2.set_ylabel('Number of gates',fontsize=16)  # we already handled the x-label with ax1
ax2.plot(x, z,'k^-', label="Number of Gates")
ax2.tick_params(axis='y')
ax2.legend(loc=6)
ax2.grid()
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = "14"
plt.xticks(x)

plt.show()
fig.savefig("qubits_iterations_gates.pdf")
```

```python
def f_gates(x, y,gate):
    NO_OF_QUBITS_FITNESS = 8
    NO_OF_QUBITS_INDIVIDUAL = x
    POPULATION_SIZE = 2**NO_OF_QUBITS_INDIVIDUAL
    NO_OF_MAX_GROVER_ITERATIONS = y#int(math.sqrt(2**NO_OF_QUBITS_INDIVIDUAL-1))
    NO_OF_QUBITS_CARRY = 2 * NO_OF_MAX_GROVER_ITERATIONS + 1
    NO_OF_GENERATIONS = 8
    
    if gate==0:
        H_GATES = NO_OF_QUBITS_INDIVIDUAL+1+NO_OF_MAX_GROVER_ITERATIONS*(2+2*NO_OF_QUBITS_FITNESS+3)
        return H_GATES
    elif gate==1:
        X_GATES = POPULATION_SIZE*2*NO_OF_QUBITS_INDIVIDUAL+NO_OF_MAX_GROVER_ITERATIONS*(2+3*NO_OF_QUBITS_FITNESS)
        return X_GATES
    elif gate == 2:
        CX_GATES = 2*NO_OF_MAX_GROVER_ITERATIONS*(2*8+2*8+1)
        return CX_GATES
    elif gate == 3:
        CCX_GATES = 2*NO_OF_MAX_GROVER_ITERATIONS*(8+8)
        return CCX_GATES
    else:
        MCT_GATES = POPULATION_SIZE*NO_OF_QUBITS_FITNESS+NO_OF_MAX_GROVER_ITERATIONS*2
        return MCT_GATES
    
q = [4, 8,12,16,20,24] #number of qubits
y = list(map(lambda x:int(math.sqrt(2**x-1)), q))
qy= list(zip(q,y))
h= list(map(lambda x:f_gates(x[0],x[1],0),qy))
x= list(map(lambda x:f_gates(x[0],x[1],1),qy))
cx= list(map(lambda x:f_gates(x[0],x[1],2),qy))
ccx= list(map(lambda x:f_gates(x[0],x[1],3),qy))
mct= list(map(lambda x:f_gates(x[0],x[1],4),qy))

print(h)
print(x)
print(cx)
print(ccx)
print(mct)
```

```python
fig, ax = plt.subplots()
ax.plot(q, h,'o--', label="Number of H gates", color = "black",linewidth=3,markersize=10)
ax.plot(q, x,'^:', label="Number of X gates" , color = "dimgray",linewidth=3,markersize=10)
ax.plot(q, cx,'s-.', label="Number of CX gates", color = "gray",linewidth=3,markersize=10)
ax.plot(q, ccx,'p-.', label="Number of CCX gates", color = "silver",linewidth=3,markersize=10)
ax.plot(q, mct,'P-', label="Number of MCT gates", color = "darkgray",linewidth=3,markersize=10)
ax.set_yscale('log')
ax.legend()
ax.grid()
ax.set_xlabel("Number of qubits in individual register",fontsize=16)
ax.set_ylabel("Number of gates[log10]",fontsize=16)
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = "14"
plt.xticks(q)
plt.show()
fig.savefig("qubits_gates.pdf")
```
